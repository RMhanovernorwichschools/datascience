Further investigation of plotting

```python
from numpy import radians, sin, linspace
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ballistics_points = [(x,(100**2)*sin(2*radians(x))/9.81) for x in linspace (0,90,200)]
```

```python
df = pd.DataFrame(ballistics_points)
df.columns=['angle','range']
```

```python
plt.plot(df.angle, df.range)
plt.xlabel(df.angle.name+' (degrees)')
plt.ylabel(df.range.name+' (meters)')
plt.legend()
plt.show()
```

```python
def range_of_deg(x):
    return (100**2)*sin(2*radians(x))/9.81
```

```python
df2 = pd.DataFrame()
df2['angle_column']=linspace(0,90,200)
df2['range_column']= range_of_deg(df2.angle_column)
```

```python
plt.plot(df2.angle_column, df2.range_column)
plt.xlabel('angle (degrees)')
plt.ylabel('range (meters)')
plt.show()
```
