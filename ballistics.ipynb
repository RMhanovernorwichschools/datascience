# Ballistics and the Range Equation
This is an investigation of the so-called *range equation*: $d=\frac {v^2} g \sin 2\theta$.

```python
from math import radians, sin, pi
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
def rangeformula(v, theta):
    '''Inputs:
    v = velocity in m/s
    theta = launch angle with respect to horizon in radians
    Returns:
    Impact distance in m
    '''
    return sin(2*theta)*v**2/9.81
```

```python
angledegrees = linspace(0,90,num=200)
```

```python
rangefromdegrees =[]
for x in angledegrees:
    rangefromdegrees.append(rangeformula(100,radians(x)))
```

```python
angleradians = linspace(0,pi/2,num=200)
```

```python
rangefromradians =[]
for x in angleradians:
    rangefromradians.append(rangeformula(100,x))
```

```python
plt.plot(angledegrees, rangefromdegrees)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
plt.plot(angleradians, rangefromradians)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (radians)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
velocities = linspace(0,200,num=200)
```

```python
rangefrom15degrees =[]
for x in velocities:
    rangefrom15degrees.append(rangeformula(x,radians(15)))
rangefrom30degrees =[]
for x in velocities:
    rangefrom30degrees.append(rangeformula(x,radians(30)))
rangefrom45degrees =[]
for x in velocities:
    rangefrom45degrees.append(rangeformula(x,radians(45)))
rangefrom60degrees =[]
for x in velocities:
    rangefrom60degrees.append(rangeformula(x,radians(60)))
rangefrom75degrees =[]
for x in velocities:
    rangefrom75degrees.append(rangeformula(x,radians(75)))
```

```python
plt.plot(velocities, rangefrom15degrees, label="15 degrees")
plt.plot(velocities, rangefrom30degrees, label="30 degrees")
plt.plot(velocities, rangefrom45degrees, label="45 degrees")
plt.plot(velocities, rangefrom60degrees, label="60 degrees")
plt.plot(velocities, rangefrom75degrees, label="75 degrees")
plt.legend()

plt.title('Range as a function of initial velocity at varying angles')
plt.xlabel('Velocity (meters per second)')
plt.ylabel('Range (meters)')
plt.show()
```
