## Interest Table
This notebook uses pandas and matplot to calculate an interest table, in which all variables are constants except time, and display the results graphically. The interest is annual, so the notebook uses the below equation to calculate final value: 

$A=P(1+r)^t$

The account starts with $1000 and the interest rate is 3.5%, so:

- P = $1000
- r = 0.035
- t = time in years (*independent variable*)
- A = resulting value in dollars (*dependent variable*)

$A=1000(1+0.035)^t$

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
inttab = pd.DataFrame()
inttab['years']=range(31)
inttab['value_float']=1000*(1.035**inttab.years)
inttab['value_str']= ['$'+str(int(round(x,0))) for x in inttab.value_float]
```

```python
inttab
```

```python
plt.plot(inttab.years,inttab.value_float)
plt.xlabel(inttab.years.name)
plt.ylabel('value ($)')
plt.show()
```
